### Importing Libraries

```python
import json
import datetime
from collections import defaultdict
```

### Read Data from "record.json" file

```python
fd = open("record.json",'r')
txt = fd.read()
fd.close()

records = json.loads(txt)
```

### Read Sales Data from "sales.json" file

```python
fd = open('sales.json', 'r')
txt1 = fd.read()
fd.close()

sales = json.loads(txt1)
```

### Products Purchased

```python
flag = True
while(flag):
    
    print()
    print("*"*46, "\nPlease Enter the Details of Purchased Products\n", "*"*46, sep = "")
    i = 1
    prod_id = input("Please enter the product id: ")
    if prod_id not in records:
        print("Please enter correct product id!!")
        continue
    qty = int(input("Please enter the product quantity: "))
    if qty > records[prod_id]['Quantity']:
        print("Sorry product out of order! Current stock quantity is", records[prod_id]['Quantity'], "units.")
        continue
    records[prod_id]['Quantity'] = records[prod_id]['Quantity'] - qty
    dic = {prod_id : qty}
    
    flag1 = True
    while(flag1):
        inp = input("Would you like to add more products (Y/N)?: ")

        if inp=='Y' or inp=='y':
            prod_id = input("Please enter the product id: ")
            if prod_id not in records:
                print("Please enter correct product id!!")
                continue
            qty = int(input("Please enter the product quantity: "))
            if qty > records[prod_id]['Quantity']:
                print("Sorry product out of order! Current stock quantity is", records[prod_id]['Quantity'], "units.")
                continue
            records[prod_id]['Quantity'] = records[prod_id]['Quantity'] - qty
            
            if prod_id not in dic:
                dic[prod_id] = qty
            else:
                dic[prod_id] += qty
            i += 1

        elif inp=='N' or inp=='n':
            flag1 = False
            flag = False
            
        else:
            print("\nPLEASE ENTER A VALID OPTION!!")
```
