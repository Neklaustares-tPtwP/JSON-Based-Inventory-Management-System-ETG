### Importing Libraries

```python
import json
```

### Read Data from "record.json" file

```python
fd = open("record.json", 'r')
txt = fd.read()
fd.close()

records = json.loads(txt)
```

```python
records
```

### Update, Add and Delete Items in Inventory

```python
flag = True
while(flag):
    
    print()
    print()
    print("*"*17, "\nSELECT AN OPTION") 
    print("*"*17)
    print("1. UPDATE ITEM\n2. ADD NEW ITEM\n3. DELETE ITEM\n4. EXIT")
    
    option = int(input("Print enter your choice (1/2/3/4): "))
    
    if option==1:
        prod_id = input("Please enter the product id: ")
        qty = int(input("Please enter the product quantity: "))
        price = int(input("Please enter the product price: "))
        records[prod_id]['Quantity'] = records[prod_id]['Quantity'] + qty
        records[prod_id]['Price'] = price
        
    elif option==2:
        prod_id = input("Enter product id: ")
        if prod_id in records:
            print("Item already present in inventory. Enter 1 to update the item!!")
            continue
        prod_name = input("Enter product name: ")
        comp_name = input("Enter company name: ")
        qty = int(input("Enter product quantity: "))
        types = input("Enter product type: ")
        price = int(input("Enter product price: "))
        records[prod_id] = {'Product': prod_name, 'Company': comp_name, 'Quantity': qty, 'Type': types, 'Price': price}
    
    elif option==3:
        prod_id = input("Enter product id: ")
        if prod_id not in records:
            print("Entered product id doesnot exist!!")
            continue
        del records[prod_id]
    
    elif option==4:
        print("Thank you!!")
        flag = False
    
    else:
        print("\nPLEASE ENTER A VALID OPTION!!")
        
```

### Update Items in "records.json" file

```python
js = json.dumps(records)

fd = open("record.json",'w')
fd.write(js)
fd.close()
```
